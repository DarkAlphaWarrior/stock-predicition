```
#import dependencies
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split

#load the boston data set from sklearn.datasets and #print it
from sklearn.datasets import load_boston
boston = load_boston()
print(boston)

#transform the data set into a data frame
#data = the data we want as the x values
#features_names=the column names of data
#target=the target var or price of the houses

df_x =pd.DataFrame(boston.data,columns=boston.feature_names)
df_y=pd.DataFrame(boston.target)

#get some statistics from the dataset,count,mean
df_x.describe()
#initialize the linear regression model
reg=linear_model.LinearRegression()
#split the data into 67% training and 33% testing data
x_train,x_test,y_train,y_test= train_test_split(df_x,df_y,test_size= 0.33,random_state=42)

#train the model with our training data
reg.fit(x_train,y_train)

#print the coeficiente/weights for each feature of our #model
print(reg.coef_)
#print the prediction on our test data
y_pred=reg.predict (x_test)
print (y_pred)

#print the actual values
print(y_test)

#check the model perfomance mse
print (np.mean((y_pred - y_test)**2))

#check the metrics
from sklearn.metrics import mean_squared_error
print (mean_squared_error(y_test, y_pred))




```
